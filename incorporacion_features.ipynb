```python
import numpy as np 
import pandas as pd 
```

```python
data = pd.read_csv('data_seleccion.csv',sep = ',', index_col= 'Unnamed: 0')
data.head(3)
```

```python
data.shape
```

```python
data['surface_total_in_m2_2'] = data['surface_total_in_m2']**2
data.head()
```

```python
#para incorporación de columna con distancia al punto de referencia
data_distance = pd.read_csv('data_distance.csv',sep = ',', index_col= 'field_1')
data_distance.head(3)
```

```python
distancia_serie = data_distance.distancia
```

```python
data = pd.concat([data,distancia_serie],axis=1)
```

```python
data.head(1)
```

```python
import re

list_busquedas = ['ambientes','habitaciones']

for busqueda in list_busquedas :
    data[busqueda] = data['description'].apply(lambda x: re.search(r'(?P<rooms>\d+)(?: {})'.format(busqueda), x, re.IGNORECASE).group('rooms') 
                                                                  if re.search(r'(?P<rooms>\d+)(?: {})'.format(busqueda), x, re.IGNORECASE) 
                                                                  else None)
```

```python
list_numeros = ['dos','tres','cuatro', 'cinco']

for busqueda in list_numeros :
    data[busqueda] = data['description'].apply(lambda x: re.search(r'(?P<rooms>{})(?: ambientes)'.format(busqueda), x, re.IGNORECASE).group('rooms') 
                                                                  if re.search(r'(?P<rooms>{})(?: ambientes)'.format(busqueda), x, re.IGNORECASE) 
                                                                  else None)
```

```python
data['monoambiente'] = data['description'].apply(lambda x: re.search(r'(?P<rooms>monoambiente)', x, re.IGNORECASE).group('rooms') 
                                                                  if re.search(r'(?P<rooms>monoambiente)', x, re.IGNORECASE) 
                                                                  else None)

#587 monoambientes
```

```python
#para testear datos en columnas
mask_ambientes = data['ambientes'].notnull()
print(data['ambientes'].notnull().sum())
print(data[mask_ambientes][['place_name','price_usd_per_m2','surface_total_in_m2','ambientes']].sample(5))

mask_habit = data['habitaciones'].notnull()
print(data['habitaciones'].notnull().sum())
print(data[mask_habit][['place_name','price_usd_per_m2','surface_total_in_m2','habitaciones']].sample(5))
```

```python
data['dos'] = data['dos'].apply(lambda x : 2 if isinstance(x,str) and x.upper() == 'DOS' else x)
data['tres'] = data['tres'].apply(lambda x : 3 if isinstance(x,str) and x.upper() == 'TRES' else x)
data['cuatro'] = data['cuatro'].apply(lambda x : 4 if isinstance(x,str) and x.upper() == 'CUATRO' else x)
data['cinco'] = data['cinco'].apply(lambda x : 5 if isinstance(x,str) and x.upper() == 'CINCO' else x)
data['monoambiente'] = data['monoambiente'].apply(lambda x : 1 if isinstance(x,str) and x.upper() == 'MONOAMBIENTE' else x)
```

```python
#para testear datos en columnas

for numero in list_numeros:
    mask = data[numero].notnull()
    print(f'Propiedades con {numero} ambientes: {str(data[numero].notnull().sum())}')

print('Propiedades con 1 ambiente: ' + str(data['monoambiente'].notnull().sum()))
```

```python
#para ver variación de cantidades antes de imputación
print(data['ambientes'].notnull().sum())

#incorporación de datos obtenidos en columna "ambientes"
for num in list_numeros:
    if data['ambientes'].isna().any():
        data['ambientes'] = data['ambientes'].fillna(data[num])

if data['ambientes'].isna().any():
    data['ambientes'] = data['ambientes'].fillna(data['monoambiente'])

#ver correcta imputación
print(data['ambientes'].notnull().sum())
```

```python
#control
numero = 'cuatro'
mask_numero = data[numero].notnull()
print(data[mask_numero][['place_name','price_usd_per_m2','surface_total_in_m2','ambientes',numero]].sample(5))
```

```python
#unificación de datos en rooms:
print('Cantidad de datos en rooms: ' + str(data['rooms'].notnull().sum()))

if data['rooms'].isna().any():
    data['rooms'] = data['rooms'].fillna(data['ambientes'])
    
print('Cantidad de datos en rooms: ' + str(data['rooms'].notnull().sum()))
```

```python
#borrado de columnas innecesarios y creada para obtención de datos
data.shape
data.drop(['dos','tres','cuatro','cinco','monoambiente','ambientes'], axis=1, inplace=True)
data.head(5)
```

```python
data.to_csv("data.csv")
```
